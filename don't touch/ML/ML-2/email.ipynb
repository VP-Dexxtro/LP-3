```python
"""Problem Statement

Classify the email using the binary classification method. Email Spam detection has two 
states: 
a) Normal State – Not Spam, 
b) Abnormal State – Spam. 
Use K-Nearest Neighbors and Support Vector Machine for classification. Analyze their performance."""
```

```python
import pandas as pd
```

```python
df = pd.read_csv("emails.csv")
```

```python
df.head()
```

```python
# We don't require first and last columns 
x = df.iloc[:,1:3001]
x
```

```python
# Target variable/feature
y = df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size = 0.25, random_state=1)
```

```python
# Support Vector Machine
from sklearn.svm import SVC
svc = SVC(kernel='linear')   
```

```python
from sklearn.metrics import accuracy_score
svc.fit(train_x,train_y)
pred1 = svc.predict(test_x)
print("Accuracy Score : ", accuracy_score(pred1,test_y))
```

```python
# k-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(train_x, train_y)
pred2 = knn.predict(test_x)
print("Accuracy Score : ", accuracy_score(pred2,test_y))
```
